```python
from f_score import *
import matplotlib.pyplot as plt
```

Data Setup

```python
file = "csi_sample_data.csv"
scorer = F_Score(file, 2782000000)
f_scores = scorer.get_f_score()
print(f_scores)
```

Top 5 T-scores

```python
top_5 = np.argsort(f_scores, axis=0)[-5:]
for idx in top_5:
    print(idx, ":" ,f_scores[idx])
max_t = f_scores[top_5[4]]
p104 = max_t + (max_t*0.04) #relative placement calculations for replaceable data sets
p103 = max_t + (max_t*0.03)
```

Graph of T-scores over channel index

```python
x = np.arange(1008)
plt.plot(x, f_scores, color="purple")
plt.vlines(507, 0, p103, linestyle="--", color="green")
plt.text(345, p104, "Amplitude/Phase Split", color="green")
for idx in top_5:
    plt.vlines(idx, 0, p103, linestyles="dotted", color="red")
plt.text(-40, p104, "Top 5 T-scores", color="red")
plt.ylabel("Fisher Score")
plt.xlabel("Channel Index")
plt.title("Fisher Scores at Every Channel")
plt.show()
```
